```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression,f_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from collections import defaultdict
from sklearn.ensemble import RandomForestRegressor

def Indices(num,X,y):
    index = range(X.shape[1])
    """
    #correlation
    selector = SelectKBest(f_regression, k=num).fit(X,y)
    indices1 = selector.get_support(indices=True)
    

    #mutual information

    selector = SelectKBest(mutual_info_regression, k=num).fit(X,y)
    indices2 = selector.get_support(indices=True)
    indices2

    #Mean decrease impurity

    #Load boston housing dataset as an example
    rf = RandomForestRegressor(criterion = "mae",n_estimators = 100)
    rf.fit(X, y)

    
    scores_impurity = sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_),index), 
                 reverse=True)
    indices3 = list(map(lambda x: x[1] ,scores_impurity[0:num]))
    #print(indices3)
    """
    #Mean decrease accuracy

    
    rf = RandomForestRegressor()
    scores = defaultdict(list)

    #crossvalidate the scores on a number of different random splits of the data
    for i in range(100):
        X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.33, random_state=42)
        r = rf.fit(X_train, Y_train)
        acc = r2_score(Y_test, rf.predict(X_test))
        for i in range(X.shape[1]):
            X_t = X_test.copy()
            np.random.shuffle(X_t[:, i])
            shuff_acc = r2_score(Y_test, rf.predict(X_t))
            scores[index[i]].append((acc-shuff_acc)/acc)
    #print("Indices determined by Mean decrease accuracy")
    #print("Features sorted by their score:")
    scores_accuracy = sorted([(round(np.mean(score), 4), i) for
                  i, score in scores.items()], reverse=True)
    indices4 = list(map(lambda x: x[1] ,scores_accuracy[0:num]))
    #print(indices4)
    #return [indices1,indices2,indices3,indices4]
    return indices4
```

```python
def get_features(dependent_features,target_features,num):
    df = raw_df[dependent_features]
    target_df = raw_df[target_features]
    #df.loc[df['DiggingActor'] == '#DIV/0!']
    #NOTE male ID 219 doesnt have Actor score so i made it zeros...
    # ideally i would remove him from the dataset...
    # however this is a prototype i am yet to give in depth analysis

    # fill missing values wit 0
    df = df.fillna(0)
    target_df = target_df.fillna(0) # change this some values had NA
    X = df.values.astype('float')
    y = target_df.values.astype('float') 
    actor = y[:,0]
    partner = y[:,1]
    actor_indices = Indices(num,X,actor)
    partner_indices = Indices(num,X,partner)
    
    actor_features = [dependent_features[i] for i in actor_indices]
    partner_features = [dependent_features[i] for i in partner_indices]
    #NOTE change this tooo
    actor_dataset = (X[:,actor_indices], actor)
    partner_dataset = (X[:,partner_indices], partner)
    
    return (actor_dataset,actor_features), (partner_dataset,partner_features)
    
```

```python
male_df = raw_df[male_features].fillna(0)
female_df = raw_df[female_features].fillna(0)
features = male_df.columns.tolist() #same as female but with F at the front

pairwise_target_df = raw_df[pairwise_target_features].fillna(0) #change this some values are NA

male_X = male_df.values.astype('float') 
female_X = female_df.values.astype('float') 
pairwise_target = pairwise_target_df.values.astype('float') 

Mrel = pairwise_target[:,0] # pairwise male relationship score
Frel = pairwise_target[:,1] # pairwise female relationship score
# pairwise difference
X = male_X-female_X

MIndcies = Indices(20,male_X,Mrel)
FIndcies = Indices(20,female_X,Frel)

Mfeatures = [features[i] for i in MIndcies]
Ffeatures = [features[i] for i in FIndcies]

pairwise_MX = (X[:,MIndcies], Mrel)
pairwise_FX = (X[:,FIndcies], Frel)

pairwise_MData = (pairwise_MX,Mfeatures), (pairwise_FX, Ffeatures )
```
